```python
import numpy as np
import math
import pandas as pd
import copy
```

## Questão 2

```python
dataset = pd.read_csv('lista4db.csv', sep=",")
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1:].values
```

```python
def apply_to_each_element(mtx, fn):
    new_mtx = copy.copy(mtx)
    shape_x = mtx.shape[0]
    shape_y = mtx.shape[1]
    for i in range(shape_x):
        for j in range(shape_y):
            new_mtx[i, j] = fn(new_mtx[i, j])
    return new_mtx
```

```python
def convert_to_float(el):
    new_el = comma_to_dot(el)
    return float(new_el)
```

```python
def comma_to_dot(str_el):
    return str_el.replace(',', '.')
```

```python
X = apply_to_each_element(X, convert_to_float)
Y = apply_to_each_element(Y, convert_to_float)
```

### 2.1

```python
mean_X = np.mean(X, axis=0)
mean_Y = np.mean(Y, axis=0)
```

```python
def dp(random_sample):
    mean = np.mean(random_sample, axis=0)
    sum_arg = [np.square(x - mean) for x in random_sample]
    num = sum(sum_arg)
    den = random_sample.shape[0] - 1
    return math.sqrt(num / den)

dp_X = np.apply_along_axis(dp, 0, X)
dp_Y = np.apply_along_axis(dp, 0, Y)
```

```python
data = [np.concatenate((mean_X, mean_Y)), np.concatenate((dp_X, dp_Y))]
pd.DataFrame(data, index=["Média", "Desvio Padrão"], columns=["X1", "X2", "X3", "X4", "Y"])
```

```python
def normalize(el, mean, dp):
    return (el - mean) / dp

norm_X1 = apply_to_each_element(X[:, 0:1], lambda el : normalize(el, mean_X[0], dp_X[0]))
norm_X2 = apply_to_each_element(X[:, 1:2], lambda el : normalize(el, mean_X[1], dp_X[1]))
norm_X3 = apply_to_each_element(X[:, 2:3], lambda el : normalize(el, mean_X[2], dp_X[2]))
norm_X4 = apply_to_each_element(X[:, 3:4], lambda el : normalize(el, mean_X[3], dp_X[3]))

norm_X = np.concatenate((norm_X1, norm_X2, norm_X3, norm_X4), axis=1)
norm_Y = apply_to_each_element(Y, lambda el : normalize(el, mean_Y[0],dp_Y[0]))

norm_data = np.concatenate((norm_X, norm_Y), axis=1)
pd.DataFrame(norm_data, columns=["X1", "X2", "X3", "X4", "Y"])
```

```python
def r(x, y):    
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    num = sum([np.mean((x[i] - mean_x) * (y[i] - mean_y)) for i in range(x.shape[0])])
    den_x = math.sqrt(sum([(el_x - mean_x) ** 2 for el_x in x]))
    den_y = math.sqrt(sum([(el_y - mean_y) ** 2 for el_y in y]))
    return num / (den_x * den_y)

values = np.concatenate((X, Y), axis=1)

def R(mtx):
    num_of_entries = mtx.shape[1]
    R = np.zeros((num_of_entries, num_of_entries), dtype=float)
    
    for i in range(num_of_entries):
        for j in range(num_of_entries):
            R[i, j] = r(mtx[:, i], mtx[:, j])
    return R

header = ["X1", "X2", "X3", "X4", "Y"]
corr = R(values)
pd.DataFrame(corr, index=header, columns=header)
```
